```python
import pandas as pd
import os
import csv
# Paths
ma_2015_path = os.path.join('../Resources', 'Massachusetts', 'ma_2015.csv')
ma_2016_path = os.path.join('../Resources', 'Massachusetts', 'ma_2016.csv')
ma_2017_path = os.path.join('../Resources', 'Massachusetts', 'ma_2017.csv')
ma_2018_path = os.path.join('../Resources', 'Massachusetts', 'ma_2018.csv')
ma_2019_path = os.path.join('../Resources', 'Massachusetts', 'ma_2019.csv')
```

```python
# Read in file
ma_2015 = pd.read_csv(ma_2015_path, sep=',', parse_dates=['arrest_date'])
ma_2016 = pd.read_csv(ma_2016_path, sep=',', parse_dates=['arrest_date'])
ma_2017 = pd.read_csv(ma_2017_path, sep=',', parse_dates=['ARREST_DATE'])
ma_2018 = pd.read_csv(ma_2018_path, sep=',', parse_dates=['ARREST_DATE'])
ma_2019 = pd.read_csv(ma_2019_path, sep=',', parse_dates=['ARREST_DATE'])
```

```python
# Convert to lower case
ma_2017.columns = ma_2017.columns.str.lower()
ma_2018.columns = ma_2018.columns.str.lower()
ma_2019.columns = ma_2019.columns.str.lower()
```

```python
# Merging
merge_one = ma_2015.merge(ma_2016, how='outer')
merge_two = merge_one.merge(ma_2017, how='outer')
merge_three = merge_two.merge(ma_2018, how='outer')
merge_total = merge_three.merge(ma_2019, how='outer')
merge_total
```

```python
offense_type = 35
merge_total = merge_total[merge_total.offense_type_id.eq(offense_type)]
merge_total.head()
```

```python
merge_total = merge_total.drop(columns=['arrestee_id', 'incident_id', 'arrestee_seq_num', 'arrest_num', 'arrest_type_id', 'multiple_indicator', 'offense_type_id', 'age_id', 'age_num', 'race_id', 'ethnicity_id', 'resident_code', 'under_18_disposition_code', 'clearance_ind', 'ff_line_number', 'age_range_low_num', 'age_range_high_num', 'data_year'])
merge_total.head()
```

```python
# Date conversion
merge_total['year'] = merge_total['arrest_date'].dt.year
merge_total['month'] = merge_total['arrest_date'].dt.month
merge_total['day'] = merge_total['arrest_date'].dt.day
merge_total.head()
```

```python
merge_total['state'] = 'MA'
massachusetts_df = merge_total.drop(columns={'arrest_date', 'day'})
massachusetts_df
massachusetts_df.to_csv('../Resources/Massachusetts/massachusetts.csv', index=False)
```
