Creating a Pivot table to make the gender be part of the column by year month and state

```python
ss = ss.rename(columns={"sex_code":"id"})
```

```python
hh = ss.pivot_table("count",["year","month","state"],"id")
```

```python
hh.head()
```

```python
crime_table = hh.reset_index()
```

```python
crime_table
```

```python
kk=crime_table.sort_values(by=["state","year","month"])
```

```python
kk.head(20)
```

```python
crime_table['F'] = crime_table['F'].fillna (0)
```

```python
crime_table['M'] = crime_table['M'].fillna (0)
```

```python
crime_table
```

```python
crime_table["number_of_crimes"] = crime_table["F"]+crime_table["M"]
```

```python
crime_table = crime_table.rename(columns={"F": "number_f", "M" :"number_m"})
```

```python
crime_table
```

```python
crime_table = crime_table[["state","year","month","number_of_crimes","number_f","number_m"]]
```

```python
crime_table = crime_table.drop('month', axis= 'columns')
```

Finishing the transforming of the crimes table and getting it ready to have the same structure as the Deaths table
by regrouping by year and state only

```python
crime_table = crime_table.groupby(["year","state"]).sum().reset_index()
```

Final query to joing the two data sources from database 

```python
pd.read_sql("""select 
overdose_deaths.name_of_state,
overdose_deaths.state,
overdose_deaths.year,
overdose_deaths.number_of_deaths,
overdose_deaths.indicator,
drug_crimes.number_of_crimes,
drug_crimes.number_f,
drug_crimes.number_m
from overdose_deaths inner join drug_crimes 
on drug_crimes.state = overdose_deaths.state and 
   drug_crimes.year = overdose_deaths.year    
   ORDER BY overdose_deaths.state, overdose_deaths.year 
   ASC;
""", connection)
```
