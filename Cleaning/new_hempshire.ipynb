```python
import pandas as pd
import csv
import os
```

```python
filepath_NH_2015 = os.path.join('../Resources', 'New Hampshire', 'NH_2015_nibrs_arrestee.csv')
filepath_NH_2016 = os.path.join('../Resources', 'New Hampshire', 'NH_2016_nibrs_arrestee.csv')
filepath_NH_2017 = os.path.join('../Resources', 'New Hampshire', 'NH_2017_NIBRS_ARRESTEE.csv')
filepath_NH_2018 = os.path.join('../Resources', 'New Hampshire', 'NH_2018_NIBRS_ARRESTEE.csv')
filepath_NH_2019 = os.path.join('../Resources', 'New Hampshire', 'NH_2019_NIBRS_ARRESTEE.csv')
```

# Files to DataFrame

```python
NH_2015_df = pd.read_csv(filepath_NH_2015)
NH_2016_df = pd.read_csv(filepath_NH_2016)
NH_2017_df = pd.read_csv(filepath_NH_2017)
NH_2018_df = pd.read_csv(filepath_NH_2018)
NH_2019_df = pd.read_csv(filepath_NH_2019)
```

# Aligning Desired Column Names

```python
NH_2017_df_2 = NH_2017_df.rename(columns={'DATA_YEAR':'year','ARREST_DATE':'arrest_date','OFFENSE_TYPE_ID':'offense_type_id','SEX_CODE':'sex_code'})
NH_2018_df_2 = NH_2018_df.rename(columns={'DATA_YEAR':'year','ARREST_DATE':'arrest_date','OFFENSE_TYPE_ID':'offense_type_id','SEX_CODE':'sex_code'})
NH_2019_df_2 = NH_2019_df.rename(columns={'DATA_YEAR':'year','ARREST_DATE':'arrest_date','OFFENSE_TYPE_ID':'offense_type_id','SEX_CODE':'sex_code'})

```

# Creating Year Column

```python
NH_2015_df['year'] = 2015
NH_2016_df['year'] = 2016
```

# Aligning Date Format

```python
merge_1 = NH_2015_df.merge(NH_2016_df, how = 'outer')

dates = merge_1['arrest_date'].str.split('-')
month = []
for number in dates:
    month.append(int(number[1]))
    
merge_1['arrest_date'] = month
merge_1 = merge_1.rename(columns={'arrest_date':'month'})
merge_1.head()
```

# Merging DataFrames Into One

```python
merge_2 = NH_2017_df_2.merge(NH_2018_df_2, how = 'outer')
merge_3 = merge_2.merge(NH_2019_df_2, how = 'outer')

date = merge_3['arrest_date'].str.split('-')
month = []
for number in date:
    if number[1] == 'JAN':
        month.append(1)
    if number[1] == 'FEB':
        month.append(2)
    if number[1] == 'MAR':
        month.append(3)
    if number[1] == 'APR':
        month.append(4)
    if number[1] == 'MAY':
        month.append(5)
    if number[1] == 'JUN':
        month.append(6)
    if number[1] == 'JUL':
        month.append(7)
    if number[1] == 'AUG':
        month.append(8)
    if number[1] == 'SEP':
        month.append(9)
    if number[1] == 'OCT':
        month.append(10)
    if number[1] == 'NOV':
        month.append(11)
    if number[1] == 'DEC':
        month.append(12)
        
merge_3['arrest_date'] = month
merge_3 = merge_3.rename(columns={'arrest_date':'month'})
merge_3.head()
```

# Final Merge and Clean

```python
merge_4 = merge_1.merge(merge_3, how = 'outer')

merge_4['state'] = 'NH'

NH_df_dirty = merge_4.loc[merge_4['offense_type_id'] == 35,['year','month','state','sex_code']]

NH_df = NH_df_dirty[['year', 'month','state','sex_code']]
NH_df.head()
```

```python
NH_df.to_csv('../Resources/New Hampshire/new_hampshire.csv')
```
