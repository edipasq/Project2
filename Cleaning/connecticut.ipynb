```python
import pandas as pd
import os
import csv
# Paths
ct_2015_path = os.path.join('../Resources', 'Connecticut', 'ct_2015.csv')
ct_2016_path = os.path.join('../Resources', 'Connecticut', 'ct_2016.csv')
ct_2017_path = os.path.join('../Resources', 'Connecticut', 'ct_2017.csv')
ct_2018_path = os.path.join('../Resources', 'Connecticut', 'ct_2018.csv')
ct_2019_path = os.path.join('../Resources', 'Connecticut', 'ct_2019.csv')
```

```python
# Read all files
ct_2015 = pd.read_csv(ct_2015_path, sep = ',', parse_dates= ['arrest_date'])
ct_2016 = pd.read_csv(ct_2016_path, sep = ',', parse_dates= ['arrest_date'])
ct_2017 = pd.read_csv(ct_2017_path, sep = ',', parse_dates= ['ARREST_DATE'])
ct_2018 = pd.read_csv(ct_2018_path, sep = ',', parse_dates= ['ARREST_DATE'])
ct_2019 = pd.read_csv(ct_2019_path, sep = ',', parse_dates= ['ARREST_DATE'])
```

```python
# Convert to lower case 
ct_2017.columns = ct_2017.columns.str.lower()
ct_2018.columns = ct_2018.columns.str.lower()
ct_2019.columns = ct_2019.columns.str.lower()
```

```python
merge_one = ct_2015.merge(ct_2016, how='outer')
merge_two = merge_one.merge(ct_2017, how='outer')
merge_three = merge_two.merge(ct_2018, how='outer')
merge_total = merge_three.merge(ct_2019, how='outer')
merge_total
```

```python
# Filter offense type
offense_type = 35
merge_total = merge_total[merge_total.offense_type_id.eq(offense_type)]
# merge_total = merge_total.reset_index(drop=True)
merge_total
```

```python
# Drop uneeded columns
merge_total = merge_total.drop(columns=['arrestee_id', 'incident_id', 'arrestee_seq_num', 'arrest_num', 'arrest_type_id', 'multiple_indicator', 'offense_type_id', 'age_id', 'age_num', 'race_id', 'ethnicity_id', 'resident_code', 'under_18_disposition_code', 'clearance_ind', 'ff_line_number', 'age_range_low_num', 'age_range_high_num', 'data_year'])
merge_total
```

```python
# Date conversion
merge_total['year'] = merge_total['arrest_date'].dt.year
merge_total['month'] = merge_total['arrest_date'].dt.month
merge_total['day'] = merge_total['arrest_date'].dt.day
merge_total
```

```python
# Drop and add
merge_total['state'] = 'CT'
connecticut_df = merge_total.drop(columns={'arrest_date', 'day'})
connecticut_df
connecticut_df.to_csv('../Resources/Connecticut/connecticut_df.csv', index=False)
```
